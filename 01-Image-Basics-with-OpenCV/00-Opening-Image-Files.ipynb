jadi kalo di matplotlib color channelnya beda sama opencv

matplotlib = RGB (Red Green Blue)

opencv = BGR (Blue Green Red)

jadi color channelnya terbalik, makanya warna gambarnya jadi aneh, harus di convert dulu

```python
plt.imshow(img)
```

```python
# ini cara convertnya
# kita mau convert dari BGR ke RBG
# jadi BGR2RGB, 2 nya itu dibaca to
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# next convert jadi grayscale
# read fotonya langsung convert
img_gray = cv2.imread('/Users/clcx/Documents/GitHub/OpenCv-Deep-Learning/Pierian/DATA/00-puppy.jpg',
cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
# warnanya aneh karena color channelnya salah lg
```

```python
# bisa kita convert
img_gray2 = cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB)
plt.imshow(img_gray2)
```

```python
# atau langsung ubah cmap dari plt nya pas convert
img_gray = cv2.imread('/Users/clcx/Documents/GitHub/OpenCv-Deep-Learning/Pierian/DATA/00-puppy.jpg',
cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

## Resize Image

```python
img_rgb.shape
# width, height, color channels
# color channels 3 artinya berwarna
# kalo grayscale cuma array 2d
```

```python
img = cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

### By ratio

```python
# contoh kita mau resize pake ratio
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img_rgb,(0,0),img_rgb,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

### Flipping Images

```python
# bisa 0,1,atau -1
```

```python
new_img_flipped = cv2.flip(new_img,0)
plt.imshow(new_img_flipped)
```

```python
new_img_flipped = cv2.flip(new_img,1)
plt.imshow(new_img_flipped)
```

```python
new_img_flipped = cv2.flip(new_img,-1)
plt.imshow(new_img_flipped)
```
