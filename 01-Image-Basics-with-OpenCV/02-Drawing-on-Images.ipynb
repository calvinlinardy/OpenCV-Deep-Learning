## Shapes

### Rectangles

* img Image.
* pt1 Vertex of the rectangle.
* pt2 Vertex of the rectangle opposite to pt1 .
* color Rectangle color or brightness (grayscale image).
* thickness Thickness of lines that make up the rectangle. Negative values, like #FILLED,mean that the function has to draw a filled rectangle.
* lineType Type of the line. See #LineTypes
* shift Number of fractional bits in the point coordinates.

```python
# kita buat rectangle
# pt1 = top left
# pt2 = bottom right
cv2.rectangle(blank_img,pt1=(385,0),pt2=(510,128),color=(0,255,0),thickness=5)
plt.imshow(blank_img)
# thicknessnya keliatan ga seimbang
# soalnya rada terlalu mojok
```

```python
cv2.rectangle(blank_img,pt1=(385,10),pt2=(500,128),color=(0,255,0),thickness=5)
plt.imshow(blank_img)
```

### Circles

```python
# tinggal tentuin center(titik tengahnya) terus radius
cv2.circle(img=blank_img,center=(100,100),radius=70,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```

### Text

```python
# org itu bottom left textnya
# selalu pake linetype = LINE_AA
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img,text='Calvin',org=(10,495),fontFace=font,fontScale=4,thickness=3,color=(255,255,255),lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

### Polygons

To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2 where ROWS are number of vertices and it should be of type int32. 


```python
# jadi ada 2 step tambahan buat polygons

# vertices itu kan posisi x,y 4 points dari polygonnya

# pertama harus reshape verticesnya ((-1,1,2))

# kedua pas di masukin ke polylines, harus dalam bentuk list
```

```python
# kita buat blank_img baru lg biar jelas
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]])
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
# udah jadi array 3d
pts
```

```python
# ini kalo isClosed nya false
cv2.polylines(blank_img,[pts],isClosed=False,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```

```python
# ini kalo isClosed nya True
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,255,0),thickness=5)
plt.imshow(blank_img)
```
